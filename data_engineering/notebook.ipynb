## 4. Fungsi Luas Segitiga

```python
import math
def luas_segitiga_alas_tinggi(alas, tinggi): 
    return alas * tinggi / 2
def luas_segitiga_sisi(a, b, c): 
    # Jika merupakan segitiga siku-siku
    if(a**2 + b**2 == c**2):
        return a*b/2
    elif(c**2 + b**2 == a**2):
        return c*b/2
    elif(a**2 + c**2 == b**2):
        return a*c/2
    # Jika bukan siku-siku â†’ gunakan rumus Heron
    s = (a + b + c) / 2
    return math.sqrt(s * (s - a) * (s - b) * (s - c))
```

```python
# case A
luas_segitiga_alas_tinggi(3, 4)
```

```python
# case B
luas_segitiga_sisi(3, 4, 5)
```

## 5. Conditional Logics

```python
# konversi nilai 
def convert_nilai(nilai):
    if nilai >= 80:
        return "A"
    elif nilai >= 60:
        return "B"
    elif nilai >= 50: 
        return "C"
    elif nilai >= 40: 
        return "D"
    return "E"
```

```python
# Contoh nilai = 55
convert_nilai(55)
```

## 6. While Loop

```python
# while loop dari nilai 20, increment 5, stop 100
nilai = 20
while nilai <= 100: 
    print("Nilai Saat ini : {}".format(nilai))
    nilai += 5
```
