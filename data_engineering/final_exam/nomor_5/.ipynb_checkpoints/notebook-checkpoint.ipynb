```python
from scipy import stats
```

```python
# coba prototype
# Menggunakan filter untuk mengambil subset data
subset_data = products_df[(products_df.product_weight_g == 300) &
                          (products_df.product_length_cm == 16.0) &
                          (products_df.product_height_cm == 7.0) &
                          (products_df.product_width_cm == 20.0)]
```

```python
subset_data["product_name_lenght"].mean()
```

```python
products_null_df = products_df[products_df.product_category_name.isna()&products_df.product_name_lenght.isna()&products_df.product_description_lenght.isna()&products_df.product_photos_qty.isna()]
```

```python
products_null_df.head()
```

```python
products_null_df['product_length_cm'].mode()[0]
```

```python
for i in products_null_df.index:
  subset_data = products_df[(products_df.product_weight_g == products_df.loc[i, "product_weight_g"]) &
                            (products_df.product_length_cm == products_df.loc[i, "product_length_cm"]) &
                            (products_df.product_height_cm == products_df.loc[i, "product_height_cm"]) &
                            (products_df.product_width_cm == products_df.loc[i, "product_width_cm"])]
  subset_data.dropna(axis=0, inplace=True)
  if not subset_data.empty:
    products_df.loc[i, "product_category_name"] = subset_data.product_category_name.value_counts().idxmax()
    products_df.loc[i, "product_name_lenght"] = subset_data["product_name_lenght"].mean()
    products_df.loc[i, "product_description_lenght"] = subset_data["product_description_lenght"].mean()
    products_df.loc[i, "product_photos_qty"] = subset_data["product_photos_qty"].mean()
```

```python
products_df[products_df.product_category_name.isna()]
```

```python
# sisanya kita drop saja
products_df.dropna(axis=0, inplace=True)
```

```python
# buat kolom simpangan estimasi
estimation_deviation_days = orders_df['order_delivered_carrier_date'] - orders_df['order_estimated_delivery_date']
estimation_deviation_days = estimation_deviation_days.apply(lambda x: x.total_seconds())
estimation_deviation = estimation_deviation_days.apply(lambda x: "On Time / Earlier" if x<= 0 else("Late"))
orders_df['estimation_deviation'] = estimation_deviation
```

### performa delivery (bagaimana ketepatan estimasi dalam history data)

```python
colors = [ "#D2D2D2", "#7D1007"]
estimationDev = all_df.groupby(by="estimation_deviation").order_id.nunique().reset_index()
estimationDev.rename(columns={
    "order_id": "order_count"
}, inplace=True)

plt.figure(figsize=(10, 5))

sns.barplot(
    y="order_count",
    x="estimation_deviation",
    data=estimationDev.sort_values(by="order_count", ascending=True),
    palette=colors
)
total_orders = estimationDev['order_count'].sum()  # hitung total orders

# Add data labels to the bars
for index, row in estimationDev.iterrows():
    percentage = row['order_count'] / total_orders * 100  # buat persentase
    label = f"{row['order_count']} ({percentage:.2f}%)"

    plt.annotate(label,
                 xy=(row.name, row['order_count']),
                 xytext=(-5, 0),
                 textcoords='offset points',
                 ha='center',
                 fontsize=12)

plt.title("Number of Order by Estimation Deviation", loc="center", fontsize=15)
plt.ylabel(None)
plt.xlabel(None)
plt.tick_params(axis='x', labelsize=12)
plt.show()
```

### RFM Analysis (Customers dengan value tertinggi pada perusahaan berdasarkan analisis RFM)

```python
# coba pakai all data
all_df = pd.read_csv('data/all_data.csv')
datetime_col = ["order_purchase_timestamp", "order_approved_at", "order_delivered_carrier_date", "order_delivered_customer_date", "order_estimated_delivery_date", "shipping_limit_date"]
```

```python
for col in datetime_col:
    all_df[col] = pd.to_datetime(all_df[col])
```

```python
all_df.info()
```

* Recency: parameter yang digunakan untuk melihat kapan terakhir seorang pelanggan melakukan transaksi.
* Frequency: parameter ini digunakan untuk mengidentifikasi seberapa sering seorang pelanggan melakukan transaksi.
* Monetary: parameter terakhir ini digunakan untuk mengidentifikasi seberapa besar revenue yang berasal dari pelanggan tersebut.

```python
rfm_df = all_df.groupby(by="customer_id", as_index = False).agg({
    "order_purchase_timestamp" : "max", # mengambil tanggal order terakhir
    "order_id" : "nunique", # menghitung jumlah order
    "price" : "sum" # menghitung jumlah revenue dihasilkan
})
```

```python
rfm_df.columns = ["customer_id", "max_order_timestamp", "frequency", "monetary"]
```

```python
rfm_df.info()
```

```python
# hitung kapan terakhir pelanggan transaksi
rfm_df["max_order_timestamp"] = rfm_df["max_order_timestamp"].dt.date
recent_date = all_df["order_purchase_timestamp"].dt.date.max()
rfm_df["recency"] = rfm_df["max_order_timestamp"].apply(lambda x : (recent_date-x).days)
```

```python
rfm_df.drop("max_order_timestamp", axis=1, inplace=True)
rfm_df.head()
```

```python
rfm_df.sort_values(by="recency", ascending = True).head()
```

### Keterbatasan Analisis
- Rentang Waktu dan Konteks Eksternal: Analisis ini terbatas pada data historis tahun 2016-2018. Tanpa data mengenai kondisi ekonomi makro Brazil saat itu, kampanye pemasaran spesifik yang dijalankan, atau aktivitas kompetitor, sulit untuk menyimpulkan kausalitas penuh di balik kenaikan atau penurunan tren penjualan tertentu.
- Bias "One-Time Buyer" pada RFM: Seperti yang ditemukan dalam analisis RFM, mayoritas pelanggan hanya melakukan satu kali transaksi (Frequency = 1). Hal ini membuat segmentasi pelanggan menjadi kurang tajam karena sulit membedakan antara pelanggan setia yang baru bergabung dengan pelanggan yang memang hanya berniat membeli sekali (churn).
- Fokus pada Revenue, Bukan Profit: Data yang tersedia hanya memuat harga jual (price) dan biaya pengiriman (freight value), namun tidak mencakup Harga Pokok Penjualan (HPP) atau biaya operasional lainnya. Oleh karena itu, analisis ini hanya mencerminkan top-line growth (pendapatan kotor) dan belum bisa menggambarkan profitabilitas bersih perusahaan (net profit).

- Analisis Sentimen (NLP) pada Ulasan Pelanggan: Mengingat adanya dataset order_reviews, eksplorasi selanjutnya akan menerapkan Natural Language Processing untuk membedah teks ulasan. Ini akan menjawab mengapa suatu produk mendapatkan rating rendah—apakah karena kualitas produk, pengiriman yang buruk, atau ketidaksesuaian deskripsi—sehingga tindakan perbaikan bisa lebih spesifik.

- Forecasting Penjualan Lanjutan: Menggunakan metode Time Series Analysis yang lebih canggih (seperti SARIMA atau Prophet) untuk memprediksi permintaan di masa depan dengan mempertimbangkan faktor musiman (seasonality) dan tren, guna membantu manajemen stok dan logistik.
